Columna EV Pricing

Entre los valores que encontramos en la columna EV Pricing:

- Gratuito: Algunas estaciones de carga ofrecen el servicio de manera gratuita. 

- Por kWh (kilovatio-hora): Este es un esquema comÃºn donde se cobra una tarifa basada en la cantidad de electricidad consumida durante la carga. Los precios pueden variar, como "$0.13 por kWh" o "$0.35 por kWh".

- Por hora: Otra estructura de precios cobra por el tiempo que el vehÃ­culo estÃ¡ conectado a la estaciÃ³n de carga, sin importar cuÃ¡nta electricidad consuma. Ejemplos incluyen "$1.50 por hora" o "$2 por hora".

- Por sesiÃ³n: Algunas estaciones cobran una tarifa plana por cada sesiÃ³n de carga, independientemente de la duraciÃ³n o la cantidad de energÃ­a consumida, como "$2 por sesiÃ³n" o "$5 por sesiÃ³n".

- Tarifas variables: Algunos esquemas de precios pueden incluir tarifas variables por estacionamiento y/o energÃ­a, indicando que el costo puede variar segÃºn ciertas condiciones, como la duraciÃ³n del estacionamiento o la cantidad de energÃ­a consumida.

- Donaciones: En algunos casos, la carga puede ser gratuita, pero se anima a los usuarios a hacer donaciones como una forma de apoyar el servicio.

- Tarifas complejas: Algunas estaciones utilizan estructuras de precios mÃ¡s complejas, combinando tarifas por kWh, por tiempo, y cargos por estacionamiento, lo que puede resultar en un cÃ¡lculo mÃ¡s complicado del costo total de la carga.

La diversidad en las estructuras de precios refleja los diferentes modelos de negocio y estrategias de las entidades que operan las estaciones de carga, lo cual dificulta encontrar un factor comÃºn para agrupar los datos.

## III Tratamiento outliers y duplicados

### 1 Outliers


### EV Level2 EVSE Num (NÃºmero de Conectores de Nivel 2 EVSE)

- **InterpretaciÃ³n:** Los valores que son considerados outliers (por ejemplo, 16) indican estaciones de carga con un nÃºmero relativamente alto de conectores de Nivel 2 EVSE disponibles. Mientras que la mayorÃ­a de las estaciones de carga pueden tener menos conectores, estos outliers representan estaciones de carga de gran capacidad, posiblemente ubicadas en Ã¡reas de alta demanda o diseÃ±adas para servir a muchos vehÃ­culos simultÃ¡neamente.

### EV DC Fast Count (Conteo de Carga RÃ¡pida DC)

- **InterpretaciÃ³n:** Los valores outliers como 20 y 25 sugieren estaciones de carga que ofrecen una cantidad significativamente alta de puntos de carga rÃ¡pida DC, mucho mÃ¡s que la media de las estaciones. Esto podrÃ­a indicar centros de carga diseÃ±ados para un trÃ¡fico muy alto o para servir como hubs principales de carga en rutas clave o Ã¡reas metropolitanas donde la demanda de carga rÃ¡pida es alta.

Los ourliers de ambas columnas resaltan estaciones de carga que son inusualmente bien equipadas para ofrecer servicios de carga, lo que no implica que haya un error en los datos.

```python
# Seleccionamos solo las columnas numÃ©ricas
df_numeric = df_ny.select_dtypes(include=['float64', 'int64'])

# Calcula el rango intercuartÃ­lico (IQR) para cada columna numÃ©rica
Q1 = df_numeric.quantile(0.25)
Q3 = df_numeric.quantile(0.75)
IQR = Q3 - Q1

# Identifica los outliers
outliers = ((df_numeric < (Q1 - 1.5 * IQR)) | (df_numeric > (Q3 + 1.5 * IQR)))

# Muestra los outliers
for column in outliers.columns:
    if outliers[column].any():
        print(f"Outliers in {column}:")
        print(df_numeric[column][outliers[column]])
        print()
```
